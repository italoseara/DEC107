```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
openmp_1 = pd.read_csv('out/openmp_1.csv', sep=',')
openmp_2 = pd.read_csv('out/openmp_2.csv', sep=',')
openmp_4 = pd.read_csv('out/openmp_4.csv', sep=',')
openmp_8 = pd.read_csv('out/openmp_8.csv', sep=',')
openmp_1.head()
```

```python
mpi_1 = pd.read_csv('out/mpi_1.csv', sep=',')
mpi_2 = pd.read_csv('out/mpi_2.csv', sep=',')
mpi_4 = pd.read_csv('out/mpi_4.csv', sep=',')
mpi_8 = pd.read_csv('out/mpi_8.csv', sep=',')
mpi_1.head()
```

```python
# Tempo de execuo x Tamanho da matriz por Threads (OpenMP)

plt.figure(figsize=(10,6))
plt.plot(openmp_1['Mat Size'], openmp_1['Mean Time (s)'], marker='o', label='Sequencial')
plt.plot(openmp_2['Mat Size'], openmp_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(openmp_4['Mat Size'], openmp_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(openmp_8['Mat Size'], openmp_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Tempo mdio de execuo (s)')
plt.title('Tempo de execuo x Tamanho da matriz por Threads (OpenMP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tempo de execuo x Tamanho da matriz por Threads (MPI)

plt.figure(figsize=(10,6))
plt.plot(mpi_1['Mat Size'], mpi_1['Mean Time (s)'], marker='o', label='Sequencial')
plt.plot(mpi_2['Mat Size'], mpi_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(mpi_4['Mat Size'], mpi_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(mpi_8['Mat Size'], mpi_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Tempo mdio de execuo (s)')
plt.title('Tempo de execuo x Tamanho da matriz por Threads (MPI)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Speedup x Tamanho da matriz por Threads (OpenMP)

# Speedup = Tempo sequencial / Tempo paralelo

plt.figure(figsize=(10,6))
plt.plot(openmp_2['Mat Size'], openmp_1['Mean Time (s)'] / openmp_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(openmp_4['Mat Size'], openmp_1['Mean Time (s)'] / openmp_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(openmp_8['Mat Size'], openmp_1['Mean Time (s)'] / openmp_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Speedup')
plt.title('Speedup x Tamanho da matriz por Threads (OpenMP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Speedup x Tamanho da matriz por Threads (MPI)

# Speedup = Tempo sequencial / Tempo paralelo

plt.figure(figsize=(10,6))
plt.plot(mpi_2['Mat Size'], mpi_1['Mean Time (s)'] / mpi_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(mpi_4['Mat Size'], mpi_1['Mean Time (s)'] / mpi_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(mpi_8['Mat Size'], mpi_1['Mean Time (s)'] / mpi_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Speedup')
plt.title('Speedup x Tamanho da matriz por Threads (MPI)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Eficincia vs Tamanho da matriz por Threads (OpenMP)

# Eficincia = Speedup / Nmero de threads

plt.figure(figsize=(10,6))
plt.plot(openmp_2['Mat Size'], (openmp_1['Mean Time (s)'] / openmp_2['Mean Time (s)']) / 2, marker='o', label='2 Threads')
plt.plot(openmp_4['Mat Size'], (openmp_1['Mean Time (s)'] / openmp_4['Mean Time (s)']) / 4, marker='o', label='4 Threads')
plt.plot(openmp_8['Mat Size'], (openmp_1['Mean Time (s)'] / openmp_8['Mean Time (s)']) / 8, marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Eficincia')
plt.title('Eficincia vs Tamanho da matriz por Threads (OpenMP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Eficincia vs Tamanho da matriz por Threads (MPI)

# Eficincia = Speedup / Nmero de threads

plt.figure(figsize=(10,6))
plt.plot(mpi_2['Mat Size'], (mpi_1['Mean Time (s)'] / mpi_2['Mean Time (s)']) / 2, marker='o', label='2 Threads')
plt.plot(mpi_4['Mat Size'], (mpi_1['Mean Time (s)'] / mpi_4['Mean Time (s)']) / 4, marker='o', label='4 Threads')
plt.plot(mpi_8['Mat Size'], (mpi_1['Mean Time (s)'] / mpi_8['Mean Time (s)']) / 8, marker='o', label='8 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Eficincia')
plt.title('Eficincia vs Tamanho da matriz por Threads (MPI)')
plt.legend()
plt.grid(True)
plt.show()
```
