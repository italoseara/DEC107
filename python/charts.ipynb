```python
# Importar bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sequencial = pd.read_csv('benchmarks/sequencial.dat', sep=',')
sequencial.head()
```

```python
parallel_2 = pd.read_csv('benchmarks/parallel_opt_2.dat', sep=',')
parallel_4 = pd.read_csv('benchmarks/parallel_opt_4.dat', sep=',')
parallel_8 = pd.read_csv('benchmarks/parallel_opt_8.dat', sep=',')
parallel_16 = pd.read_csv('benchmarks/parallel_opt_16.dat', sep=',')

# parallel_2 = pd.read_csv('benchmarks/parallel_simple_2.dat', sep=',')
# parallel_4 = pd.read_csv('benchmarks/parallel_simple_4.dat', sep=',')
# parallel_8 = pd.read_csv('benchmarks/parallel_simple_8.dat', sep=',')
# parallel_16 = pd.read_csv('benchmarks/parallel_simple_16.dat', sep=',')
```

```python
# Tempo de execu√ß√£o vs Tamanho da matriz por Threads

plt.figure(figsize=(10,6))
plt.plot(sequencial['Mat Size'], sequencial['Mean Time (s)'], marker='o', label='Sequencial')
plt.plot(parallel_2['Mat Size'], parallel_2['Mean Time (s)'], marker='o', label='Paralelo (2 threads)')
plt.plot(parallel_4['Mat Size'], parallel_4['Mean Time (s)'], marker='o', label='Paralelo (4 threads)')
plt.plot(parallel_8['Mat Size'], parallel_8['Mean Time (s)'], marker='o', label='Paralelo (8 threads)')
plt.plot(parallel_16['Mat Size'], parallel_16['Mean Time (s)'], marker='o', label='Paralelo (16 threads)')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Tempo m√©dio de execu√ß√£o (s)')
plt.title('Tempo de execu√ß√£o vs Tamanho da matriz por Threads')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Speedup vs Tamanho da matriz por Threads

# Speedup = Tempo sequencial / Tempo paralelo

plt.figure(figsize=(10,6))
plt.plot(parallel_2['Mat Size'], sequencial['Mean Time (s)'] / parallel_2['Mean Time (s)'], marker='o', label='Speedup (2 threads)')
plt.plot(parallel_4['Mat Size'], sequencial['Mean Time (s)'] / parallel_4['Mean Time (s)'], marker='o', label='Speedup (4 threads)')
plt.plot(parallel_8['Mat Size'], sequencial['Mean Time (s)'] / parallel_8['Mean Time (s)'], marker='o', label='Speedup (8 threads)')
plt.plot(parallel_16['Mat Size'], sequencial['Mean Time (s)'] / parallel_16['Mean Time (s)'], marker='o', label='Speedup (16 threads)')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Speedup')
plt.title('Speedup vs Tamanho da matriz por Threads')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Efici√™ncia vs Tamanho da matriz por Threads

# Efici√™ncia = Speedup / N√∫mero de threads

plt.figure(figsize=(10,6))
plt.plot(parallel_2['Mat Size'], (sequencial['Mean Time (s)'] / parallel_2['Mean Time (s)']) / 2, marker='o', label='Efici√™ncia (2 threads)')
plt.plot(parallel_4['Mat Size'], (sequencial['Mean Time (s)'] / parallel_4['Mean Time (s)']) / 4, marker='o', label='Efici√™ncia (4 threads)')
plt.plot(parallel_8['Mat Size'], (sequencial['Mean Time (s)'] / parallel_8['Mean Time (s)']) / 8, marker='o', label='Efici√™ncia (8 threads)')
plt.plot(parallel_16['Mat Size'], (sequencial['Mean Time (s)'] / parallel_16['Mean Time (s)']) / 16, marker='o', label='Efici√™ncia (16 threads)')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Efici√™ncia')
plt.title('Efici√™ncia vs Tamanho da matriz por Threads')
plt.legend()
plt.grid(True)
plt.show()
```
