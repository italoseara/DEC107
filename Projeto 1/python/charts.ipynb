```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sequencial = pd.read_csv('benchmarks/sequencial.dat', sep=',')
sequencial.head()
```

```python
sequencial_opt = pd.read_csv('benchmarks/sequencial_opt.dat', sep=',')
sequencial_opt.head()
```

```python
parallel_2 = pd.read_csv('benchmarks/parallel_2.dat', sep=',')
parallel_4 = pd.read_csv('benchmarks/parallel_4.dat', sep=',')
parallel_8 = pd.read_csv('benchmarks/parallel_8.dat', sep=',')
parallel_16 = pd.read_csv('benchmarks/parallel_16.dat', sep=',')
parallel_2.head()
```

```python
parallel_opt_2 = pd.read_csv('benchmarks/parallel_opt_2.dat', sep=',')
parallel_opt_4 = pd.read_csv('benchmarks/parallel_opt_4.dat', sep=',')
parallel_opt_8 = pd.read_csv('benchmarks/parallel_opt_8.dat', sep=',')
parallel_opt_16 = pd.read_csv('benchmarks/parallel_opt_16.dat', sep=',')
parallel_opt_2.head()
```

```python
# Tempo de execuo vs Tamanho da matriz por Threads (No otimizado)

plt.figure(figsize=(10,6))
plt.plot(sequencial['Mat Size'], sequencial['Mean Time (s)'], marker='o', label='Sequencial')
plt.plot(parallel_2['Mat Size'], parallel_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(parallel_4['Mat Size'], parallel_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(parallel_8['Mat Size'], parallel_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.plot(parallel_16['Mat Size'], parallel_16['Mean Time (s)'], marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Tempo mdio de execuo (s)')
plt.title('Tempo de execuo vs Tamanho da matriz por Threads (No otimizado)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tempo de execuo vs Tamanho da matriz por Threads (Otimizado)

plt.figure(figsize=(10,6))
plt.plot(sequencial_opt['Mat Size'], sequencial_opt['Mean Time (s)'], marker='o', label='Sequencial')
plt.plot(parallel_opt_2['Mat Size'], parallel_opt_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(parallel_opt_4['Mat Size'], parallel_opt_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(parallel_opt_8['Mat Size'], parallel_opt_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.plot(parallel_opt_16['Mat Size'], parallel_opt_16['Mean Time (s)'], marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Tempo mdio de execuo (s)')
plt.title('Tempo de execuo vs Tamanho da matriz por Threads (Otimizado)')
plt.grid(True)
plt.legend()
```

```python
# Speedup vs Tamanho da matriz por Threads (No otimizado)

# Speedup = Tempo sequencial / Tempo paralelo

plt.figure(figsize=(10,6))
plt.plot(parallel_2['Mat Size'], sequencial['Mean Time (s)'] / parallel_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(parallel_4['Mat Size'], sequencial['Mean Time (s)'] / parallel_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(parallel_8['Mat Size'], sequencial['Mean Time (s)'] / parallel_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.plot(parallel_16['Mat Size'], sequencial['Mean Time (s)'] / parallel_16['Mean Time (s)'], marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Speedup')
plt.title('Speedup vs Tamanho da matriz por Threads (No otimizado)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Speedup vs Tamanho da matriz por Threads (Otimizado)

# Speedup = Tempo sequencial / Tempo paralelo

plt.figure(figsize=(10,6))
plt.plot(parallel_opt_2['Mat Size'], sequencial_opt['Mean Time (s)'] / parallel_opt_2['Mean Time (s)'], marker='o', label='2 Threads')
plt.plot(parallel_opt_4['Mat Size'], sequencial_opt['Mean Time (s)'] / parallel_opt_4['Mean Time (s)'], marker='o', label='4 Threads')
plt.plot(parallel_opt_8['Mat Size'], sequencial_opt['Mean Time (s)'] / parallel_opt_8['Mean Time (s)'], marker='o', label='8 Threads')
plt.plot(parallel_opt_16['Mat Size'], sequencial_opt['Mean Time (s)'] / parallel_opt_16['Mean Time (s)'], marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Speedup')
plt.title('Speedup vs Tamanho da matriz por Threads (Otimizado)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Eficincia vs Tamanho da matriz por Threads (No otimizado)

# Eficincia = Speedup / Nmero de threads

plt.figure(figsize=(10,6))
plt.plot(parallel_2['Mat Size'], (sequencial['Mean Time (s)'] / parallel_2['Mean Time (s)']) / 2, marker='o', label='2 Threads')
plt.plot(parallel_4['Mat Size'], (sequencial['Mean Time (s)'] / parallel_4['Mean Time (s)']) / 4, marker='o', label='4 Threads')
plt.plot(parallel_8['Mat Size'], (sequencial['Mean Time (s)'] / parallel_8['Mean Time (s)']) / 8, marker='o', label='8 Threads')
plt.plot(parallel_16['Mat Size'], (sequencial['Mean Time (s)'] / parallel_16['Mean Time (s)']) / 16, marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Eficincia')
plt.title('Eficincia vs Tamanho da matriz por Threads (No otimizado)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Eficincia vs Tamanho da matriz por Threads (Otimizado)

# Eficincia = Speedup / Nmero de threads

plt.figure(figsize=(10,6))
plt.plot(parallel_opt_2['Mat Size'], (sequencial_opt['Mean Time (s)'] / parallel_opt_2['Mean Time (s)']) / 2, marker='o', label='2 Threads')
plt.plot(parallel_opt_4['Mat Size'], (sequencial_opt['Mean Time (s)'] / parallel_opt_4['Mean Time (s)']) / 4, marker='o', label='4 Threads')
plt.plot(parallel_opt_8['Mat Size'], (sequencial_opt['Mean Time (s)'] / parallel_opt_8['Mean Time (s)']) / 8, marker='o', label='8 Threads')
plt.plot(parallel_opt_16['Mat Size'], (sequencial_opt['Mean Time (s)'] / parallel_opt_16['Mean Time (s)']) / 16, marker='o', label='16 Threads')
plt.xlabel('Tamanho da matriz (N x N)')
plt.ylabel('Eficincia')
plt.title('Eficincia vs Tamanho da matriz por Threads (Otimizado)')
plt.legend()
plt.grid(True)
plt.show()
```
